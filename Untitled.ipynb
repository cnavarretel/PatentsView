```python
import pandas as pd
import glob
from pandas.io.json import json_normalize
```

```python
df = pd.concat([pd.read_json(file) for file in glob.glob("data_temp/*.json")])
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.patents.apply(pd.Series)
```

```python
df.head()
```

```python
items = []
for patent in df.itertuples():
    for assignee in patent.assignees:
        for cpc in patent.cpcs:
            items.append({
                "assignee_country": assignee["assignee_country"],
                "assignee_state": assignee["assignee_state"],
                "cpc_group_id": cpc["cpc_group_id"],
                "patent_number": patent.patent_number,
                "patent_year": patent.patent_year
            })
```

```python
df_assignees = pd.DataFrame(items).drop_duplicates().dropna()
```

```python
df_assignees.shape
```

```python
df_assignees.groupby(["assignee_state", "cpc_group_id", "patent_year", "assignee_country"]).count().reset_index()
```

```python
df_assignees.to_pickle("./us_patents.pkl")
```
